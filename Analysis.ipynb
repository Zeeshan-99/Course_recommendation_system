```python
import pandas as pd
import numpy as np
import neattext.functions as nfx
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# from dashboard import   getsubjectsperlevel, yearwiseprofit
```

```python
df= pd.read_csv('course_data.csv')
df.head()
```

```python
df.columns
```

```python
df['clean_title']=df['course_title'].apply(nfx.remove_stopwords)
df['clean_title']=df['course_title'].apply(nfx.remove_special_characters)
df.head()
```

# Vectorise the clean Title

```python
countvect= CountVectorizer()
cvmat= countvect.fit_transform(df['clean_title'])
cvmat
```

```python
print(cvmat)
```

# Cosine Similarity

```python
cos_sim= cosine_similarity(cvmat)
cos_sim
```

```python
cos_sim.shape
```
